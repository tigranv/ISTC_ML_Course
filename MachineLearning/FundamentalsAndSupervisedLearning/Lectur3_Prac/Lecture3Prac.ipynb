```python
from sklearn.preprocessing import PolynomialFeatures
from numpy.linalg import inv


def fit_linear_regression(X, Y, Lambda=0, degree=1):
    poly = PolynomialFeatures(degree)
    X = poly.fit_transform(X)
    # Identity matrix (number of parameters is the dimension)
    Identity = np.identity(len(X[1,:]))   
    # We don't add penalty to intercept
    Identity[0,0] = 0
    
    return inv(X.transpose().dot(X) + Lambda*Identity).dot(X.transpose()).dot(Y), X

def mean_square_loss(X, Y, b):

    return np.mean((X.dot(b) - Y) ** 2) ** (0.5)
```
