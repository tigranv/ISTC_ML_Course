```python
import math
import matplotlib.pyplot as plt
import random
import numpy as np 
```

### Function for countur plots

```python
def C_Plot(func, value_range, x_range, y_range):    
    X,Y= np.meshgrid(x_range, y_range)
    Z = func(X,Y)
    plt.figure(figsize=(10,10))
    
    for i in value_range:
        CS = plt.contour(X,Y,Z, [i])
        plt.clabel(CS,inline=True, fontsize=10)
    plt.show()
```

```python
def f1(x, y):
    return x*y
def f2(x, y):
    return y - x*x
def f3(x, y):
    return y*y - x*x
```

```python
C_Plot(f1, np.arange(0, 1, 0.2), np.arange(-2,2,0.02), np.arange(-2,2,0.02))
C_Plot(f2, np.arange(0, 1, 0.2), np.arange(-2,2,0.02), np.arange(-2,2,0.02))
C_Plot(f3, np.arange(0, 1, 0.2), np.arange(-2,2,0.02), np.arange(-2,2,0.02))
```
