### Bar plot

```python
n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x,y in zip(X,Y1):
    plt.text(x+0.4, y+0.05, '%.2f' % y, ha='center', va= 'bottom')

for x,y in zip(X,Y2):
    plt.text(x+0.4, -y-0.05, '%.2f' % y, ha='center', va= 'top')

plt.xticks([])
plt.yticks([])

plt.xlim(-.5,n)
plt.ylim(-1.25,+1.25)

plt.show()
```

### Contour Plots

```python
def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

plt.contourf(X, Y, f(X,Y), 8, alpha=.75, cmap='jet')
plt.colorbar()

C = plt.contour(X, Y, f(X,Y), 8, colors='black')
plt.show()
```

### Imshow

```python
def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 10
x = np.linspace(-3,3,int(3.5*n))
y = np.linspace(-3,3,int(3.0*n))
X,Y = np.meshgrid(x,y)
Z = f(X,Y)

plt.axes([0.025,0.025,0.95,0.95])
plt.imshow(Z,interpolation='nearest', cmap='gray', origin='lower')
plt.colorbar()

plt.xticks([]), plt.yticks([])
plt.show()
```

### Histgram plots

```python
from scipy.stats import norm

mu, sigma = 100, 15
x = mu + sigma*np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, facecolor='green', alpha=0.75, normed=True)

# add a 'best fit' line
y = norm.pdf(bins, mu, sigma)
l = plt.plot(bins, y, 'r-', linewidth=2)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=100,\ \sigma=15$')
plt.grid()

plt.show()
```

```python
# normal distribution center at x=0 and y=5
x = np.random.randn(100000)
y = np.random.randn(100000) + 5

plt.hist2d(x, y, bins=40, 
#            norm=matplotlib.colors.LogNorm()
          )
plt.colorbar()
plt.show()
```

### Save figure

```python
from mpl_toolkits.mplot3d import Axes3D

def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

plt.figure(figsize=plt.figaspect(0.5))

plt.subplot(121)
n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

plt.contourf(X, Y, f(X,Y), 8, alpha=.75, cmap='jet')
plt.colorbar()
C = plt.contour(X, Y, f(X,Y), 8, colors='black')

ax = plt.subplot(122, projection="3d")

ax.plot_surface(X, Y, f(X, Y))

plt.savefig("3dplot.png", dpi=300)

plt.show()
```
